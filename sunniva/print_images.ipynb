```python
import pydicom
from pydicom import dcmread
import numpy as np
import matplotlib.pylab as plt
image_dcm = ['20_0_1.dcm','20_0_10.dcm','20_1_2.dcm','20_3_17.dcm']
image_npy= ['cropped_20_0_1.npy','cropped_20_0_10.npy','cropped_20_1_2.npy','cropped_20_3_23.npy']

# specify your image path
image_path = '20_0_1.dcm'
ds = dcmread(image_path)

plt.imshow(ds.pixel_array, cmap='binary')
plt.figure()

plt.imshow(np.load('cropped_20_0_1.npy'), 'gray')
```

```python
image_list =  ['20_0_1.dcm','20_0_10.dcm','20_1_2.dcm','20_3_23.dcm']

f = plt.figure(1)
f.set_figwidth(20)
#f.set_figheight(5)
for i,item in enumerate(image_list):
    img = dcmread(item)
    plt.subplot(1,4, i+1)
    plt.imshow(img.pixel_array, 'gray')
    plt.axis('off')
    plt.title('original, {}'.format(i+1),fontsize=26)
    #plt.savefig("image{}.png".format(i+4)) 
    

plt.savefig("originale.png")

plt.show()
```

```python
f1 = plt.figure(2)
f1.set_figwidth(20)
for i,item in enumerate(image_npy):
    plt.subplot(1,4, i+1)
    plt.imshow(np.load(item), 'gray')
    plt.axis('off')
    plt.title('beskjært, {}'.format(i+1),fontsize=26)
    

plt.savefig("beskjaerte.png")        
plt.show()
```
