```python
from pydicom import dcmread
import numpy as np
import matplotlib.pylab as plt

def print_single_image(img, save=False,name=None):
    plt.figure(dpi=800)
    print('Shape: ', np.shape(np.load(img)))
    plt.imshow(np.load(img), 'gray')
    plt.axis('off')
    if save == True:
        plt.savefig(name,transparent=True,pad_inches=0,bbox_inches='tight')        
    plt.show()


def print_dcm_images(im_lst, save=False):
    """Print 4 images in a row from a list of dicom-images."""
    f = plt.figure()
    f.set_figwidth(20)
    for i,item in enumerate(im_lst):
        img = dcmread(item)
        print('Shape: ', np.shape(img.pixel_array))
        plt.subplot(1,4, i+1)
        plt.imshow(img.pixel_array, 'gray',alpha=1)
        plt.axis('off')
        plt.title('original, {}'.format(i+1),fontsize=26)
        
    if save == True:
        plt.savefig("originale.png",transparent=True,pad_inches=0,bbox_inches='tight')

    plt.show()

def print_npy_images(im_lst,save=False):
    """Print 4 images in a row from a list of .npy-images."""
    f1 = plt.figure(2)
    f1.set_figwidth(20)
    for i,item in enumerate(im_lst):
        print('Shape: ', np.shape(np.load(item)))
        plt.subplot(1,4, i+1)
        plt.imshow(np.load(item), 'gray')
        plt.axis('off')
        plt.title('beskj√¶rt, {}'.format(i+1),fontsize=26)
        
    if save == True:
        plt.savefig("beskjaerte.png",transparent=True,pad_inches=0,bbox_inches='tight')        
    plt.show()
```
